```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
# Carga el dataset de trabajo
df_netflix = pd.read_csv('datos/netflix_final.csv', index_col=0)
df_netflix
```

#### Ejercicio 1: ¿Cuáles son todos los títulos lanzados en 2021 que son del tipo 'Movie'?

```python
condicion_2021 = df_netflix['release_year'] == 2021
condicion_movie = df_netflix['type'] == 'Movie'

df_netflix[condicion_2021 & condicion_movie]
```

#### Ejercicio 2: ¿Qué títulos tienen una duración de 90 minutos? Muestra solo las columnas de título y de duración.

```python
condicion_90min = df_netflix['duration'] == '90 min'
df_90min = df_netflix[condicion_90min]
df_90min[['title','duration']]
```

#### Ejercicio 3: ¿Cuáles son todos los títulos del país 'India' lanzados después de 2018?

```python
condicion_2018 = df_netflix['release_year'] > 2018
condicion_india = df_netflix['country'] == 'India'

df_netflix[condicion_2018 & condicion_india]
```

#### Ejercicio 4: ¿Qué títulos tienen clasificación 'TV-14' o 'R'?

```python
condicion_tv14 = df_netflix['rating'] == 'TV-14'
condicion_r = df_netflix['rating'] == 'R'

df_netflix[condicion_tv14 | condicion_r]
```

#### Ejercicio 5: ¿Cuáles son los títulos dirigidos por 'Christopher Nolan' o 'Martin Scorsese'? ¿Cuántos a dirigido cada uno de los directores?

```python
condicion_nolan = df_netflix['director'] == 'Christopher Nolan'
df_netflix[condicion_nolan]
```

```python
condicion_scorsese = df_netflix['director'] == 'Martin Scorsese'
df_netflix[condicion_scorsese]
```

```python
df_netflix[condicion_nolan | condicion_scorsese]
```

#### Ejercicio 6: ¿Qué títulos pertenecen a los países 'United States', 'United Kingdom' o 'Canada'? Dentro de estos países, ¿cuántas películas tienen más de un lenguaje?

```python
condicion_usa = df_netflix['country'] == 'United States'
condicion_uk = df_netflix['country'] == 'United Kingdom'
condicion_canada = df_netflix['country'] == 'Canada'

df_paises = df_netflix[condicion_canada | condicion_uk | condicion_usa]
df_paises
```

#### Ejercicio 7: ¿Qué títulos fueron lanzados entre 2015 y 2020, inclusive?

```python
condicion_2015 = df_netflix['release_year'] >= 2015
condicion_2020 = df_netflix['release_year'] <= 2020

df_netflix[condicion_2015 & condicion_2020]
```

#### Ejercicio 8: ¿Cuáles son los títulos cuya puntuación en IMDb está entre 6.5 y 8.0 y que no sepamos cual es su duración?

```python
condicion_between_imdb = df_netflix['IMDB Score'].between(6.5,8)
condicion_duracion_nan = df_netflix['duration'].isna()

df_netflix[condicion_between_imdb & condicion_duracion_nan]
```

#### Ejercicio 9: ¿Cuáles son las películas cuya duración de 90 y 150 minutos?

```python
condicion_150min = df_netflix['duration'] == '150 min'

df_netflix[condicion_150min | condicion_90min]
```

#### Ejercicio 10: ¿Cuáles son los títulos que contienen la palabra 'Love' en su título y que sean de tipo TV Show?

```python
condicion_love = df_netflix['title'].str.contains('Love', case=False, na=False)
condidicon_tvshow = df_netflix['type'] == 'TV Show'

df_netflix[condicion_love & condidicon_tvshow]
```

#### Ejercicio 11: ¿Cuáles son los títulos en los que el director tiene el apellido 'Johnson'?

```python
condicion_johnson = df_netflix['director'].str.contains('Johnson', case=False, na=False)
df_netflix[condicion_johnson]
```

#### Ejercicio 12: ¿Tenemos algún documental ('Documentaries') de Mike Tyson?

```python
condicion_documentaries = df_netflix['listed_in'].str.contains('Documentaries', case=False, na=False)
condicion_tyson = df_netflix['title'].str.contains('Mike Tyson', case=False, na=False)

df_netflix[condicion_documentaries & condicion_tyson]
```
